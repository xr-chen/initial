## 复现鸢尾花分类

此处我们采用Sequential函数搭建顺序网络结构

```python
import tensorflow as tf
from sklearn import datasets
import numpy as np

x_train=datasets.load_iris().data#输入特征
y_train=datasets.load_iris().target#标签

np.random.seed(116)
np.random.shuffle(x_train)
np.random.seed(116)
np.random.shuffle(y_train)
tf.random.set_seed(116)
```

然后在Sequential中搭建模型,定义神经元个数，激活函数，正则化方法

```python
model=tf.keras.models.Sequential([
    tf.keras.layers.Dense(3,activation='softmax',kernel_regularizer=tf.keras.regularizers.l2())##神经元个数，激活函数，正则化方法
])
```

选择优化器，配置训练方法，数据集中给的标签是0，1，2不是概率分布，而神经网络的输出是概率分布，所以选择SparseCategoricalCrossentropy函数作为loss函数

```python
model.compile(optimizer=tf.keras.optimizers.SGD(lr=0.1),
             loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),#激活函数是softmax所以不是原始输出选false
             metrics=['sparse_categorical_accuracy'])#同理
```

在fit中执行训练过程

```python
model.fit(x_train,y_train,batch_size=32,epochs=500,validation_split=0.2,validation_freq=20)#选择作为测试集的比例和测试的频率

model.summary()
```

## 含有跳连的非顺序网络结构

使用Sequential可以搭建含有顺序结构的神经网络，但是无法搭建含有跳连的非顺序网络结构此处用clss类来展示如何搭建，这一部分代码与原来相同

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras import Model
from sklearn import datasets
import numpy as np

x_train=datasets.load_iris().data#输入特征
y_train=datasets.load_iris().target#标签

np.random.seed(116)
np.random.shuffle(x_train)
np.random.seed(116)
np.random.shuffle(y_train)
tf.random.set_seed(116)
```

然后定义鸢尾花模型的类

```python
class IrisModel(Model):
    def __init__(self):#初始化参数，定义网络层数
        super(IrisModel,self).__init__()
        self.d1=Dense(3,activation='softmax',kernel_regularizer=tf.keras.regularizers.l2())
    def call(self,x):#实现前向传播
        y=self.d1(x)
        return y
model=IrisModel()
```

其余部分也相同

```python
model.compile(optimizer=tf.keras.optimizers.SGD(lr=0.1),
             loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),#激活函数是softmax所以不是原始输出选false
             metrics=['sparse_categorical_accuracy'])#同理
model.fit(x_train,y_train,batch_size=32,epochs=500,validation_split=0.2,validation_freq=20)#选择作为测试集的比例和测试的频率

model.summary()
```

## MNIST数据集

这里采用Sequential实现手写数字识别的训练

```python
import tensorflow as tf

mnist=tf.keras.datasets.mnist
(x_train,y_train),(x_test,y_test)=mnist.load_data('C:\\Users\\xr chen\\Downloads\\mnist.npz')
x_train,x_test=x_train/255.0,x_test/255#转化到0，1区间便于神经网络吸收

model=tf.keras.models.Sequential([
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128,activation='relu'),
    tf.keras.layers.Dense(10,activation='softmax')
])

model.compile(optimizer='adam',
             loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
             metrics=['sparse_categorical_accuracy'])

model.fit(x_train,y_train,batch_size=32,epochs=5,validation_data=(x_test,y_test),validation_freq=1)
model.summary()
```

同样我们再次采用类来定义神经网络模型

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense,Flatten
from tensorflow.keras import Model

mnist=tf.keras.datasets.mnist
(x_train,y_train),(x_test,y_test)=mnist.load_data('C:\\Users\\xr chen\\Downloads\\mnist.npz')
x_train,x_test=x_train/255.0,x_test/255#转化到0，1区间便于神经网络吸收
```

```python
class MnistModel(Model):
    def __init__(self):
        super(MnistModel,self).__init__()
        self.flatten=Flatten()
        self.d1=Dense(128,activation='relu')
        self.d2=Dense(10,activation='softmax')
    def call(self,x):
        x=self.flatten(x)
        x=self.d1(x)
        y=self.d2(x)
        return y
```

```python
model=MnistModel()
model.compile(optimizer='adam',
             loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
             metrics=['sparse_categorical_accuracy'])
model.fit(x_train,y_train,batch_size=32,epochs=5,validation_data=(x_test,y_test),validation_freq=1)
model.summary()
```
